```python
import pandas as pd
```

```python
df = pd.read_csv("bioinf-jobs-dataset.csv")



# removing samples that have no salary estimate
df = df[df["Salary Estimate"] != "-1"]

# salary parsing
salary = df["Salary Estimate"].apply(lambda x: x.split("(")[0])
minus_kd = salary.apply(lambda x: x.replace("K", "").replace("$",""))
df["Min_salary"] = minus_kd.apply(lambda x: int(x.split("-")[0]))
df["Max_salary"] = minus_kd.apply(lambda x: int(x.split("-")[1]))
df["Average_salary"] = (df.Min_salary + df.Max_salary)/2

# reordering columns
min_salary = df.pop("Min_salary")
df.insert(1, "Min_salary", min_salary)
min_salary = df.pop("Max_salary")
df.insert(2, "Max_salary", min_salary)
min_salary = df.pop("Average_salary")
df.insert(3, "Average_salary", min_salary)

#Company name: remove the rating numbers
df["company_txt"] = df.apply(lambda x: x["Company Name"] if x["Rating"] < 0 else x["Company Name"][:-3], axis = 1)

#TODO: Location: state and continent field


# looking if company name is in the same location as in headquarters
df["same_state"] = df.apply(lambda x: 1 if x["Company Name"] == x["Headquarters"] else 0, axis = 1)


#Founded: replace the founded year with how old the company is
df["age"] = df["Founded"].apply(lambda x: x if x < 1 else 2020 - x)


#Job description: extract core keywords (e.g. python, rna-analysis)
# other keywords 
df["python_yn"] = df["Job Description"].apply(lambda x: 1 if "python" in x.lower() else 0)
df.python_yn.value_counts()

df["java_yn"] = df["Job Description"].apply(lambda x: 1 if "java" in x.lower() else 0)
df.java_yn.value_counts()

df["bash_yn"] = df["Job Description"].apply(lambda x: 1 if "bash" in x.lower() else 0)
df.bash_yn.value_counts()

df["R_yn"] = df["Job Description"].apply(lambda x: 1 if " R " in x else 0)
df.R_yn.value_counts()

# saving the cleaned df
df.to_csv("bioinf_jobs_cleaned.csv", index = False)

```
